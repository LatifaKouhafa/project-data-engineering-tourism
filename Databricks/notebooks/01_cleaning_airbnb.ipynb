```python
import pandas as pd

df = pd.read_csv('airbnb-listings.csv', sep =';', nrows=100)
```

```python
import csv

# Chemin vers votre fichier CSV
csv_file_path = 'airbnb-listings.csv'

# Obtenir le nombre de lignes
with open(csv_file_path, 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    num_rows = sum(1 for row in reader) - 1  # Soustraire 1 pour l'en-tête
    print(f"Nombre de lignes : {num_rows}")

# Afficher la forme du CSV
print(f"Shape du CSV : ({num_rows})")
```

```python
# Lire le fichier CSV d'origine pour obtenir les colonnes
df_sample = pd.read_csv(csv_file_path, nrows=0)
print(df_sample.columns.tolist())
```

```python

# Colonnes pour le fichier non structuré (après vérification et ajustement)
non_structurees_colonnes = [
    'Listing Url', 'Name', 'Summary', 'Space', 'Description', 'Neighborhood Overview',
    'Notes', 'Transit', 'Access', 'Interaction', 'House Rules', 'Thumbnail Url',
    'Medium Url', 'Picture Url', 'XL Picture Url', 'Host URL', 'Host Name',
    'Host Since', 'Host Location', 'Host About', 'Host Thumbnail Url', 'Host Picture Url',
    'Host Neighbourhood', 'Street', 'Neighbourhood', 'Market', 'Smart Location',
    'Country', 'Amenities', 'License', 'Jurisdiction Names', 'Geolocation', 'Features'
]

# Taille des morceaux (chunks) à lire
chunk_size = 10000

# Lire le fichier CSV par morceaux et écrire dans les fichiers de sortie
for chunk in pd.read_csv(csv_file_path, chunksize=chunk_size):
    # Vérifier et ajuster les colonnes disponibles dans le chunk actuel
    available_non_structurees_colonnes = [col for col in non_structurees_colonnes if col in chunk.columns]
    
    # Séparer les colonnes non structurées
    df_non_structurees_chunk = chunk[available_non_structurees_colonnes]
    df_structurees_chunk = chunk.drop(columns=available_non_structurees_colonnes)
    
    # Ajouter les morceaux aux fichiers CSV de sortie
    df_non_structurees_chunk.to_csv('airbnb_listings_non_structurees.csv', mode='a', header=not os.path.exists('airbnb_listings_non_structurees.csv'), index=False)
    df_structurees_chunk.to_csv('airbnb_listings_structurees.csv', mode='a', header=not os.path.exists('airbnb_listings_structurees.csv'), index=False)

print("Les fichiers CSV volumineux ont été séparés et enregistrés avec succès.")
```
